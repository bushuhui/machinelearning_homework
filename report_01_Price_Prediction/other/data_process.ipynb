```python
# split train/test data
import random

file = open('lianjia.csv')
lns = file.readlines()
fout = open('lianjia_out.csv', "wt")
for l in lns:
    if len(l.strip()) == 0:
        continue
    else:
        fout.write(l.strip()+"\n")
fout.close()

lns = open('lianjia_out.csv').readlines()
n_lns = len(lns)
lns_idx = list(range(1,n_lns))
random.shuffle(lns_idx)
n_train = int(len(lns_idx)*0.7)
idx_train = lns_idx[:n_train]
idx_test  = lns_idx[n_train:]

ftrain = open('train_data.csv', "wt")
ftrain.write(lns[0].strip() + "\n")
for i in idx_train:
    ftrain.write(lns[i].strip() + "\n")
ftrain.close()

ftest = open('test_data.csv', "wt")
ftest.write(lns[0].strip() + "\n")
for i in idx_test:
    ftest.write(lns[i].strip() + "\n")
ftest.close()
```
